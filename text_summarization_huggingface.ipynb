```python
# extractive In this approach, the most important sentences or phrases from the original text are selected and combined to form a summary.
# abstractive  In this approach, a new summary is generated by understanding the context of the original text and generating new phrases and sentences that summarize its content.
```

```python
from transformers import pipeline 
import textwrap
import numpy as np 
from pprint import pprint
```

```python
doc = df[df.labels == 'business']['text'].sample(random_state=42)
```

```python
# text wrapping function

def wrap(x):
    return textwrap.fill(x, replace_whitespace = False , fix_sentence_endings = True )
```

```python
print(wrap(doc.iloc[0]))
```

```python
# now we will create the pipeline
```

```python
summarizer = pipeline('summarization')
```

```python
summarizer(doc.iloc[0].split('\n',1)[1])
```
